utils.py:
是读取data， 并且整理数据格式；
function:
parse_index_file: 传入的参数是filename. 将filename里的文件打开，每行都去除空格并加入到index[]的list里面。

sample_mask(): 传入的参数是index，和l.
创建一个mask， mask的大小为L的矩阵，并初始化为0，然后当有index的地方设置值为1.

load_data(): 传入的参数是dataset_str(dataset name). 传出数据是整理过的数据(training/test data).
输入的数据：
dataset_str.x: training instances 的特征值。
dataset_str.tx: tx instances 的特征值。
dataset_str.allx: labeled and unlabeled training instances 的特征值。
dataset_str.y: one-hot labels of the labeled training的值;
dataset_str.ty => the one-hot labels of the test instances的值;
dataset_str.ally => the labels for instances in ind.dataset_str.allx as numpy.ndarray object的值;
dataset_str.graph：index_of_neighbor_nodes 的值。
dataset_str.test.index：图中测试实例的索引，用于归纳设置为列表对象。
